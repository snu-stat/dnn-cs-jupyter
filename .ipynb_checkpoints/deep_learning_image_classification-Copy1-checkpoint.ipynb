        - Optimization method : Stochastic Gradient Descent (SGD)

## Introductoin
### 딥 러닝의 훈련(Training)이 어려운 이유
* Vanishing gradient
    - 딥 러닝은 기본적으로 back-propagation 알고리즘을 이용해 훈련
    - 제일 끝단에서 첫단으로 정보를 전달하는 과정에서 정보를 점점 잃어버리게 됨
    
* Overfitting
    - 뉴럴 네트워크(neural net)는 데이터에 비해 파라미터가 더 많은 모형
    - 훈련 데이터의 노이즈까지 모두 적합시켜 일반화가 잘 안됨
    - 딥 뉴럴 네트워크(deep neural net)는 이런 현상이 더 심각함

* 해결책
    1. convolutional layers 등의 차원 축소 방법
    1. 더욱 강력한 안정화 방법 (dropout, convolutional layers 등)
    1. sigmoid 뉴런을 대체할 rectified linear units 등의 뉴런 모델
    1. GPU 등의 병렬화 기법

### Convolutional layer

* 커널(kernel)과 입력 이미지의 컨볼루션(convolution)을 통해 feature를 얻는 과정
* 컨볼루션 필터(convolution filter)들을 각각 전체 이미지에 적용
* 네트워크 연결의 가중치 공유 $\rightarrow$ 파라미터 수를 줄임
* 이미지의 모든 부분에 동일한 필터 적용
* j, k번째 은닉 뉴런
$$a^1_{j,k}=\sigma\left(b+\sum_{l=0}^4\sum_{m=0}^4 \omega_{l,m} * a^0_{j+l,k+l}\right)$$
    + $*$ : 컨볼루션 연산자
    
* feature는 위치에 관계없이 검출됨 (이미지의 이동불변성(translation invariance)을 활용)

(http://cs231n.github.io/convolutional-networks/#conv)

### Pooling layer (sub-sampling layer)

|Pooling layer downsamples the volume spatially|Max Pooling|
|---|---|
|![](http://cs231n.github.io/assets/cnn/pool.jpeg)|![](http://cs231n.github.io/assets/cnn/maxpool.jpeg)|
(http://cs231n.github.io/convolutional-networks/#conv)

* Convolution layer으로부터 나온 출력을 단순화 (차원 축소)
* Max-pooling : 입력 영역의 최대값
* L2 pooling : 입력 영역의 제곱합의 제곱근

### Fully-connected layer
* Convolutional layer에서 얻어진 국소적 정보를 종합해 좀 더 추상적인 내용을 학습
* Dropout 안정화 방법
    + 네트워크에서 은닉 계산 단위를 무작위로 (그리고 임시적으로) 삭제
    + 모형이 노이즈에 강건하도록 보장하는 하나의 방법


### Final layer
* softmax 함수
    - 입력이 class L에 속할 확률 ($L = 1, 2, ...$)
   $$a_j^L=\frac{e^{z_j^L}}{\sum_ke^{z_k^L}}$$

### 딥 러닝을 위한 Numpy 데이터 처리

* Numpy 기초

````python
from keras.utils import np_utils
import numpy as np

a = np.array([0,0,1,1,2,2,1,1,3,3,0,0])
b = np.array([(0,1,2),(1,3,0)])

print('a')
print('type=',type(a), ', shape=', a.shape, ', ndim=', a.ndim)
print('data type=', a.dtype.name)
print('item size=', a.itemsize, ', size=', a.size)

print('b')
print('type=',type(b), ', shape=', b.shape, ', ndim=', b.ndim)
print('data type=', b.dtype.name)
print('item size=', b.itemsize, ', size=', b.size)
````

````python
a = a.reshape(6,2)
b = b.reshape(6)

a = a.astype('float32')
c = np_utils.to_categorical(b, 4)

print('a')
print('type=',type(a), ', shape=', a.shape, ', ndim=', a.ndim)
print('data type=', a.dtype.name)
print('item size=', a.itemsize, ', size=', a.size)

print('c')
print('type=',type(c), ', shape=', c.shape, ', ndim=', c.ndim)
print('data type=', c.dtype.name)
print('item size=', c.itemsize, ', size=', c.size)
````

````python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD

model = Sequential()
model.add(Dense(4, kernel_initializer='uniform', input_shape=(2,), activation='softmax'))
model.compile(loss='mse', optimizer='sgd', metrics=['accuracy'])

model.fit(a, c, batch_size=1, epochs=3, verbose=1)
````

    * 미분 자동화 (프로그래밍 감소)
* 역할 분담
    * 사람: 생성 공식을 주고, 조각들을 합침

        * CPU/GPU 계산을 같은 코드로 할 수 있음

    * Python, NumPy, SciPy, MatPlotLib

#### Compilation

 * __표준화(canonicalization)__: symbolic 표현을 간단하게 함
 * __수치 안정화(stabilization)__: 수치적 안정성을 향상시킴
    + 예. $\log(1+\exp(x))$
 * __특수화(specialization)__: 더 빠른 계산 방법을 선택
     + 예. $pow(x,2) \rightarrow sqr(x)$, 행렬 곱셈 $\rightarrow$ BLAS GEMM, 전치/부분 텐서 $\rightarrow$ 상수 시간 메모리 에일리어싱
     + 만약 GPU가 사용되면, GPU에 적합한 것으로 선택
 * __GPU 변환__: GPU 메모리로 `float32`형 데이터를 보냄
 * __코드 생성__: 파이썬 모듈 동적 컴파일

 #### Theano의 4 단계 사용 방법

1. 입력 타입을 선언
1. Symbolic 계산 : 수식 분석
1. 컴파일 : C/CUDA 코드를 만들고 컴파일
1. 계산 : 실제로 함수를 계산하고 모델을 훈련

### Tensorflow 기반의 High level API __TF Learn(tf.contrib.learn)__ 기초

````python
import tensorflow as tf
import numpy as np

# Create 100 phony x, y data points in NumPy, y = x * 0.1 + 0.3
x_data = np.random.rand(100).astype(np.float32)
y_data = x_data * 0.1 + 0.3

# Try to find values for W and b that compute y_data = W * x_data + b
# (We know that W should be 0.1 and b 0.3, but Tensorflow will
# figure that out for us.)
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b

# Minimize the mean squared errors.
loss = tf.reduce_mean(tf.square(y - y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)

# Before starting, initialize the variables.  We will 'run' this first.
init = tf.global_variables_initializer()

# Launch the graph.
sess = tf.Session()
sess.run(init)

# Fit the line.
for step in range(201):
    sess.run(train)
    if step % 20 == 0:
        print('step \t{} \t- W : {}, b : {}'.format(step, sess.run(W), sess.run(b)))
````

* Learns best fit is W: [0.1], b: [0.3]

### 이미지 분류 문제를 위한 CNN 네트워크를 구성하고 MNIST 데이터를 이용해 훈련 및 저장하기

* 훈련을 위한 데이터 준비하기

````python
import numpy as np
import json

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras import backend as K
import matplotlib.pyplot as plt
%matplotlib inline 

np.random.seed(1337)  # for reproducibility

# the data, shuffled and split between tran and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()

if K.image_dim_ordering() == 'th':
    X_train = X_train.reshape(X_train.shape[0], 1, 28, 28)
    X_test = X_test.reshape(X_test.shape[0], 1, 28, 28)
    input_shape = (1, 28, 28)
else:
    X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
    X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
    input_shape = (28, 28, 1)
    
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices
Y_train = np_utils.to_categorical(y_train, num_classes=10)
Y_test = np_utils.to_categorical(y_test, num_classes=10)

print('\nX_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
````

* 이미지 분류 문제를 위한 CNN 네트워크를 구성하기

````python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(5,5), padding='valid', input_shape=input_shape))
model.add(Activation('relu'))
model.add(Conv2D(filters=16, kernel_size=(5,5), padding='valid'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
````

````python
model.summary()
````

* 네트워크 컴파일 하기

````python
model.compile(loss='categorical_crossentropy', 
              optimizer='sgd',
              metrics=['accuracy'])
````

* 네트워크 훈련하기

````python
hist = model.fit(X_train, Y_train, batch_size=32, epochs=10, verbose=1, validation_split=0.2)
````

* 테스트 데이터로 훈련된 네트워크의 성능 확인

````python
score = model.evaluate(X_test, Y_test, verbose=1)
print('Test score:', score[0])
print('Test accuracy:', score[1])
````

* 훈련된 네트워크의 model, weight, history 저장하기

````python
# Save the model
json_string = model.to_json()
with open('mnist_model_architecture.json','w') as f:
    f.write(json_string)

# Save the weights
model.save_weights('mnist_model_weights.h5')

# Save History
with open('mnist_model_history.json','w') as fp:
    json.dump(hist.history, fp)
````

* History 확인 하기

````python
# Plot history
hist = json.loads(open('mnist_model_history.json').read())

plt.figure('history')
plt.subplot(211)
plt.title('Loss over epochs')
plt.plot(hist['loss'],'r',label='loss')
plt.plot(hist['val_loss'], 'b',label='val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.subplot(212)
plt.title('Accuracy over epochs')
plt.plot(hist['acc'],'r',label='acc')
plt.plot(hist['val_acc'], 'b',label='val_acc')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
````

### MNIST 데이터를 이용해 훈련한 네트워크 불러오기

* 훈련된 네트워크의 성능 확인을 위한 테스트 데이터 준비하기

````python
from keras.datasets import mnist
from keras.models import Sequential

from keras.models import model_from_json
from keras import backend as K
import json
import matplotlib.pyplot as plt
%matplotlib inline 

# the data, shuffled and split between tran and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()

if K.image_dim_ordering() == 'th':
    X_test = X_test.reshape(X_test.shape[0], 1, 28, 28)
    input_shape = (1, 28, 28)
else:
    X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
    input_shape = (28, 28, 1)

X_test = X_test.astype('float32')
X_test /= 255
print(X_test.shape[0], 'test samples')
````

* 훈련한 네트워크 불러오기

````python
# Reconstruct model
model = model_from_json(open('mnist_model_architecture.json').read())
model.load_weights('mnist_model_weights.h5')
model.summary()
````

* Test 결과 확인하기

````python
print('Predict classes')
classes = model.predict_classes(X_test, batch_size=32)

print('\nPredicted class : ', classes)
````

````python
print('Predict probability')
proba = model.predict_proba(X_test, batch_size=32)

print('\nPredicted probabitlity : ', proba)
````

* 각 층의 Activation 확인 하기

````python
# Get output of each layer
get_1st_layer_output = K.function([model.layers[0].input,
                                   K.learning_phase()],
                                  [model.layers[0].output])
get_2nd_layer_output = K.function([model.layers[0].input,
                                   K.learning_phase()],
                                  [model.layers[2].output])
get_3rd_layer_output = K.function([model.layers[0].input, 
                                   K.learning_phase()],
                                  [model.layers[7].output])
get_last_layer_output = K.function([model.layers[0].input,
                                   K.learning_phase()],
                                  [model.layers[10].output])
print('X_test image shape:', X_test.shape)
layer_1_output = get_1st_layer_output([X_test,0])[0]
print('Print 1st layer output', layer_1_output.shape)
layer_2_output = get_2nd_layer_output([X_test,0])[0]
print('Print 2nd layer output', layer_2_output.shape)
layer_3_output = get_3rd_layer_output([X_test,0])[0]
print('Print 3rd layer output', layer_3_output.shape)
layer_last_output = get_last_layer_output([X_test,0])[0]
print('Print last layer output', layer_last_output.shape)
````

````python
def plotvalue(index):
    plt.figure('Input data and 1~4 layer output value of X_test[{idx}]'.format(idx=index), 
               figsize=(9,6))
    plt.subplot2grid((5,6),(0,0),rowspan=2,colspan=2)
    plt.title('Input data')
    plt.imshow(X_test[index][:,:,0], cmap='bone', interpolation='nearest')
    plt.subplot2grid((5,6),(0,2))
    plt.imshow(layer_1_output[index][:,:,0], cmap='bone', interpolation='nearest')
    plt.subplot2grid((5,6),(0,3))
    plt.imshow(layer_1_output[index][:,:,1], cmap='bone', interpolation='nearest')
    plt.subplot2grid((5,6),(1,2))
    plt.imshow(layer_1_output[index][:,:,2], cmap='bone', interpolation='nearest')

    plt.subplot2grid((5,6),(0,4))
    plt.imshow(layer_2_output[index][:,:,0], cmap='bone', interpolation='nearest')
    plt.subplot2grid((5,6),(0,5))
    plt.imshow(layer_2_output[index][:,:,1], cmap='bone', interpolation='nearest')
    plt.subplot2grid((5,6),(1,4))
    plt.imshow(layer_2_output[index][:,:,2], cmap='bone', interpolation='nearest')

    plt.subplot2grid((5,6),(2,0),colspan=6)
    plt.imshow(layer_3_output[index].reshape(1,layer_3_output.shape[1]), cmap='bone', interpolation='nearest')
    
    plt.subplot2grid((5,6),(3,0),colspan=6)
    plt.title('probability prediction')
    plt.plot(proba[index],'r',label='probability')

    plt.subplot2grid((5,6),(4,0),colspan=6)
    plt.title('Last layer output={ans}, y_test={sol}'.format(ans=classes[index],sol=y_test[index]))
    plt.imshow(layer_last_output[index].reshape(1,layer_last_output.shape[1]), 
               cmap='bone', interpolation='nearest')
    
    plt.tight_layout()
    plt.show()
````

## CIFAR10 데이터를 위한 deep convolutional network

### CIFAR-10 dataset 
<img src="https://kaggle2.blob.core.windows.net/competitions/kaggle/3649/media/cifar-10.png" alt="Drawing" style="width: 250px;"/>
(https://www.kaggle.com/c/cifar-10)

* [CIFAR-10 dataset](http://keras.io/datasets/#cifar10-small-image-classification) : `[Sample_size, Channels, Rows, Cols]`
    - Channels = RGB (각 `0~255`), Rows = 32, Cols=32
    - Sample_size : 50000개의 training set, 10000개의 test set

### 이미지 분류 문제를 위한 CNN 네트워크를 구성하고 CIFAR10 데이터를 이용해 훈련 및 저장하기

````python
import numpy as np
import json

from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.utils import np_utils, generic_utils
from keras import backend as K

np.random.seed(1337)  # for reproducibility

# the data, shuffled and split between train and test sets
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

if K.image_dim_ordering() == 'th':
    X_train = X_train.reshape(X_train.shape[0], 3, 32, 32)
    X_test = X_test.reshape(X_test.shape[0], 3, 32, 32)
    input_shape = (3, 32, 32)
else:
    X_train = X_train.reshape(X_train.shape[0], 32, 32, 3)
    X_test = X_test.reshape(X_test.shape[0], 32, 32, 3)
    input_shape = (32, 32, 3)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)
````

````python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(5,5), padding='valid', input_shape=input_shape))
model.add(Activation('relu'))
model.add(Conv2D(filters=16, kernel_size=(5,5), padding='valid'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
````

````python
model.summary()
````

````python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
````

````python
hist = model.fit(X_train, Y_train, batch_size=32, epochs=2, verbose=1, validation_split=0.2)
````

````python
score = model.evaluate(X_test, Y_test, verbose=1)
print('\nTest score:', score[0])
print('Test accuracy:', score[1])
````

````python
# Save the model
json_string = model.to_json()
with open('cifar10_model_architecture.json','w') as f:
    f.write(json_string)

# Save the weights
model.save_weights('cifar10_model_weights.h5')

# Save History
with open('cifar10_model_history.json','w') as fp:
    json.dump(hist.history, fp)
````

````python
# Plot history 
hist = json.loads(open('cifar10_model_history.json').read())

plt.figure('history')
plt.subplot(211)
plt.title('Loss over epochs')
plt.plot(hist['loss'],'r',label='loss')
plt.plot(hist['val_loss'], 'b',label='val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.subplot(212)
plt.title('Accuracy over epochs')
plt.plot(hist['acc'],'r',label='acc')
plt.plot(hist['val_acc'], 'b',label='val_acc')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
````

## Transfer Learning


* 미리 훈련된 네트워크를 불러와서 Fine-tuning 하기
* Keras가 제공하는 이미지 분류를 위한 미리 훈련된 네트워크 (참고: https://keras.io/applications/)
    - Xception
    - VGG16
    - VGG19
    - ResNet50
    - InceptionV3
    - MobileNet

###  Imagenet dataset으로 미리 훈련된 VGG16 네트워크를 이용한 Fine-tune

(https://keras.io/applications/)

````python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from keras import backend as K

# create the base pre-trained model
base_model = VGG16(weights='imagenet', include_top=False)

base_model.summary()
````

* 새로운 데이터의 분류를 위해 마지막 단 추가하기

````python
# add a global spatial average pooling layer
x = base_model.output
x = GlobalAveragePooling2D()(x)
# and add a logistic layer -- let's say we have 10 classes
predictions = Dense(10, activation='softmax')(x)

# this is the model we will train
model = Model(inputs=base_model.input, outputs=predictions)

model.summary()
````

* 마지막 단만 훈련 가능하도록 미리 훈련된 InceptionV3 네트워크의 weight 고정하기

````python
# first: train only the top layers (which were randomly initialized)
# i.e. freeze all convolutional InceptionV3 layers
for layer in base_model.layers:
    layer.trainable = False
````

````python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
````

````python
model.summary()
````

````python
hist = model.fit(X_train, Y_train, batch_size=32, epochs=2, verbose=1, validation_split=0.2)
````

````python
score = model.evaluate(X_test, Y_test, verbose=1)
print('\nTest score:', score[0])
print('Test accuracy:', score[1])
````

````python
plt.figure('history')
plt.subplot(211)
plt.title('Loss over epochs')
plt.plot(hist.history['loss'],'r',label='loss')
plt.plot(hist.history['val_loss'], 'b',label='val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.subplot(212)
plt.title('Accuracy over epochs')
plt.plot(hist.history['acc'],'r',label='acc')
plt.plot(hist.history['val_acc'], 'b',label='val_acc')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
````
